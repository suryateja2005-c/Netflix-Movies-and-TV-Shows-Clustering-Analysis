## ***5. Hypothesis Testing***

### Based on your chart experiments, define three hypothetical statements from the dataset. In the next three questions, perform hypothesis testing to obtain final conclusion about the statements through your code and statistical testing.

Based on the exploratory data analysis and visualization experiments conducted on the Netflix Movies and TV Shows dataset, three meaningful hypothetical statements were formulated to validate patterns observed in the charts using statistical hypothesis testing. These hypotheses aim to convert visual insights into statistically supported conclusions that can guide data-driven business decisions.

The first hypothesis examines whether there is a significant difference in the release year distribution between Movies and TV Shows on Netflix. Visual analysis from the “Content Released Over the Years” and “Movies vs TV Shows Over Time” charts indicated that TV Shows have increased more rapidly in recent years compared to Movies. To statistically validate this observation, an Independent Two-Sample T-Test was performed, as release year is a numerical variable and content type (Movie or TV Show) represents two independent groups. The null hypothesis stated that there is no significant difference in the release year distribution between Movies and TV Shows, while the alternative hypothesis stated that a significant difference exists. The obtained p-value was less than the standard significance level of 0.05, leading to the rejection of the null hypothesis. This confirms that TV Shows are generally released more recently than Movies, reflecting Netflix’s strategic shift toward episodic content. From a business perspective, this insight supports Netflix’s focus on TV Shows to enhance long-term user engagement and subscription retention.

The second hypothesis investigates whether content rating is dependent on content type. Insights from multiple charts, including the Ratings Distribution and Content Type vs Rating visualizations, revealed that TV Shows dominate mature ratings such as TV-MA and TV-14, while Movies are more evenly distributed across rating categories. To test this relationship, a Chi-Square Test of Independence was applied, as both content type and rating are categorical variables. The null hypothesis assumed that content rating is independent of content type, whereas the alternative hypothesis assumed a dependency between the two variables. The statistical test produced a p-value less than 0.05, resulting in the rejection of the null hypothesis. This confirms that content rating is significantly dependent on whether the content is a Movie or a TV Show. From a business standpoint, this insight is valuable for audience segmentation, personalized recommendations, and age-appropriate content planning, helping Netflix balance mature and family-friendly offerings more effectively.

The third hypothesis focuses on whether movie duration significantly varies across different release years. Observations from the Movie Duration Distribution chart suggested that most movies fall within a consistent duration range of approximately 80 to 120 minutes, regardless of release year. To verify this pattern statistically, a Pearson Correlation Test was conducted, as both movie duration and release year are numerical variables. The null hypothesis stated that movie duration does not significantly vary with release year, while the alternative hypothesis suggested a significant relationship between the two variables. The resulting p-value was greater than 0.05, indicating a weak and statistically insignificant correlation. Therefore, the null hypothesis could not be rejected. This result implies that movie duration has remained relatively stable over time. From a business perspective, this stability allows Netflix to standardize movie length expectations and reduces uncertainty in content production and acquisition strategies.

In conclusion, hypothesis testing successfully validated the visual insights derived from the exploratory data analysis. The results confirm Netflix’s strategic emphasis on TV Shows, demonstrate a strong relationship between content type and rating, and show that movie duration remains consistent across release years. These statistically supported findings strengthen the reliability of the analysis and highlight how data-driven insights can inform effective content strategy, personalization, and business decision-making for streaming platforms.

### 4. Textual Data Preprocessing
(It's mandatory for textual dataset i.e., NLP, Sentiment Analysis, Text Clustering etc.)

#### 1. Expand Contraction

````python
# Expand Contraction
df['description'] = df['description'].str.replace("n't", " not", regex=True)
df['description'] = df['description'].str.replace("'re", " are", regex=True)
````

#### 2. Lower Casing

````python
# Lower Casing
df['description'] = df['description'].str.lower()
````

#### 3. Removing Punctuations

````python
# Remove Punctuations
import string

df['description'] = df['description'].str.translate(
    str.maketrans('', '', string.punctuation)
)
````

#### 4. Removing URLs & Removing words and digits contain digits.

````python
# Remove URLs & Remove words and digits contain digits
import re

df['description'] = df['description'].apply(
    lambda x: re.sub(r'http\S+|\d+', '', x)
)
````

#### 5. Removing Stopwords & Removing White spaces

````python
# Remove Stopwords
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))
````

````python
# Remove White spaces
df['description'] = df['description'].apply(
    lambda x: " ".join([word for word in x.split() if word not in stop_words])
)

# Remove extra whitespaces
df['description'] = df['description'].apply(lambda x: ' '.join(x.split()))
````

#### 6. Rephrase Text

````python
# Rephrase Text
````

#### 7. Tokenization

````python
# Tokenization
df['tokens'] = df['description'].apply(lambda x: x.split())
````

#### 8. Text Normalization

````python
# Normalizing Text (i.e., Stemming, Lemmatization etc.)
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()

df['tokens'] = df['tokens'].apply(
    lambda x: [lemmatizer.lemmatize(word) for word in x]
)
````

##### Which text normalization technique have you used and why?

Lemmatization was chosen over stemming because it preserves meaningful root words, resulting in better semantic interpretation and improved clustering accuracy.

#### 9. Part of speech tagging

````python
# POS Taging
nltk.download('averaged_perceptron_tagger_eng')
df['pos_tags'] = df['tokens'].apply(nltk.pos_tag)
````

#### 10. Text Vectorization

````python
# Vectorizing Text
from sklearn.feature_extraction.text import TfidfVectorizer

# Convert tokens back to text
df['lemmatized_text'] = df['tokens'].apply(lambda x: ' '.join(x))

# TF-IDF Vectorization
tfidf_vectorizer = TfidfVectorizer(max_features=5000)
X_tfidf = tfidf_vectorizer.fit_transform(df['lemmatized_text'])

X_tfidf.shape
````

##### Which text vectorization technique have you used and why?

TF-IDF vectorization was used because it captures both word frequency and importance while reducing the influence of common words. It is highly effective for text similarity and clustering tasks.

### 4. Feature Manipulation & Selection

#### 1. Feature Manipulation

````python
# Manipulate Features to minimize feature correlation and create new features
# Feature Manipulation

# Combine important textual fields to enrich semantic meaning
df['combined_text'] = (
    df['title'].fillna('') + ' ' +
    df['listed_in'].fillna('') + ' ' +
    df['description'].fillna('')
)

# Use combined_text for further NLP tasks
df['combined_text'] = df['combined_text'].str.lower()
````

#### 2. Feature Selection

````python
# Select your features wisely to avoid overfitting
# Feature Selection

# TF-IDF inherently performs feature selection using max_features
tfidf_vectorizer = TfidfVectorizer(max_features=5000)

X_tfidf = tfidf_vectorizer.fit_transform(df['combined_text'])
````

##### What all feature selection methods have you used  and why?

Explicit feature selection techniques such as Recursive Feature Elimination or SelectKBest were not used because the dataset primarily consists of unstructured text. Instead, TF-IDF vectorization was used as an implicit feature selection method by limiting the number of features using the max_features parameter. This ensures that only the most informative and frequently occurring words are retained, reducing noise, dimensionality, and overfitting.

##### Which all features you found important and why?

Textual features derived from content descriptions, genres, and titles were found to be the most important for this project. These features directly describe the theme, storyline, and category of Netflix content, which are critical for identifying similarity between movies and TV shows. After TF-IDF vectorization, words related to genres (such as drama, crime, romance, documentary) played a key role in defining cluster boundaries.

````python
import gradio as gr
import joblib
import re
import string
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

# Download NLTK resources
nltk.download('stopwords')
nltk.download('wordnet')

# ----------------------------
# Load trained components
# ----------------------------
kmeans_model = joblib.load("best_kmeans_model.pkl")
pca = joblib.load("pca_transformer.pkl")
tfidf = joblib.load("tfidf_vectorizer.pkl")

# ----------------------------
# Text preprocessing function
# ----------------------------
def preprocess_text(text):
    text = text.lower()
    text = text.translate(str.maketrans('', '', string.punctuation))
    text = re.sub(r'\d+|http\S+', '', text)
    stop_words = set(stopwords.words('english'))
    tokens = [w for w in text.split() if w not in stop_words]
    lemmatizer = WordNetLemmatizer()
    tokens = [lemmatizer.lemmatize(w) for w in tokens]
    return ' '.join(tokens)

# ----------------------------
# Cluster label mapping
# (adjust names if needed)
# ----------------------------
cluster_names = {
    0: "Family & Animated Content",
    1: "Crime & Thriller",
    2: "Romance & Drama",
    3: "Horror & Suspense",
    4: "Documentary & International"
}

# ----------------------------
# Prediction function
# ----------------------------
def predict_cluster(description):
    processed = preprocess_text(description)
    tfidf_vec = tfidf.transform([processed])
    pca_vec = pca.transform(tfidf_vec.toarray())
    cluster = kmeans_model.predict(pca_vec)[0]

    return f"Predicted Cluster: {cluster_names.get(cluster, 'Unknown Category')}"

# ----------------------------
# Gradio UI
# ----------------------------
interface = gr.Interface(
    fn=predict_cluster,
    inputs=gr.Textbox(
        lines=6,
        placeholder="Enter a Netflix movie or TV show description here...",
        label="Netflix Content Description"
    ),
    outputs=gr.Textbox(label="Prediction Result"),
    title="🎬 Netflix Content Clustering System",
    description=(
        "This tool uses a trained K-Means clustering model to categorize "
        "Netflix movies and TV shows based on their description."
    ),
    theme="soft"
)

interface.launch()
````
